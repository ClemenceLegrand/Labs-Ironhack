```python
# I create a dataframe with aggregating by unique customers with groupby
uniq_customers=orders.groupby(by="CustomerID").sum()
uniq_customers
```

```python
customer_spent = pd.DataFrame(uniq_customers)
customer_spent = pd.DataFrame(customer_spent["amount_spent"].sort_values(ascending=False))
customer_spent
```

```python
#use the pd.cut to identify the preferred and VIP clients / choose a qcut 
quantilerange = 4
labels = ["occasionnal","normal","preferred","VIP"]
labelled_customers = pd.qcut(customer_spent["amount_spent"],quantilerange,labels=labels,retbins=True)
labelled_customers
```

```python
label_cat= pd.DataFrame(labelled_customers).transpose()
label_cat= label_cat.rename(columns={"amount_spent":"category"})
label_cat["Amount Spent"]=customer_spent["amount_spent"]
label_cat
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# join the label_cat dataframe with the customer_spent dataframe
vip_country = label_cat.merge(orders, on ='CustomerID', how='left')
vip_country

#we use a new group by to not count each client several times, we use this method to display also the country column 

vip_country = vip_country.groupby(by="CustomerID").agg({
    'amount_spent': 'sum',
    'category':'first',
    'Country': 'first'
})

vip_country
```

```python
#we select only the VIPs
VIP = vip_country.loc[vip_country["category"] == "VIP"]
VIP
```

```python
#we group by and count them by countries
country_number = VIP.groupby("Country")["Country"].count()
country_number.sort_values(ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
#we select both categories instead of only VIP 
VIP_or_pref = vip_country.loc[(vip_country["category"] == "VIP")|(vip_country["category"] == "preferred")]
VIP_or_pref
```

```python
# we use a pivot table 
VIP_or_pref = VIP_or_pref.pivot_table(index=["Country"], columns=["category"],aggfunc='count',margins=True,margins_name="Total")
VIP_or_pref
```

```python
#we sort it
VIP_or_pref.sort_values(by=("amount_spent","Total"),ascending=False) #we specify the multi level index
```
