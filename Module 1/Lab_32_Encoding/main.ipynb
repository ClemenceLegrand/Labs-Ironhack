```python
import pandas as pd

data = {'TransactionID': [1, 2, 3, 4, 5],
        'ClientID': [231, 765, 453, 231, 892],
        'Profession': ['Self-employed', 'students', 'Horeca', 'self-employed', 'finance'],
        'Bank_dep': ["009", "005", "007", "009", "003"],
        'Risk': ['High', 'high', 'medium', 'high', 'low'],
        'Number of credits': [2, 3, 5, 2, 3],
        'Revenue': [30200, 12700, 89400, 30200, 740000]}

df = pd.DataFrame(data)
df
```

```python
#inutile
categorical_df = df.loc[:,["TransactionID","Profession","Bank_dep","Risk","Number of credits"]]
categorical_df
```

```python
#Dummies method
encode_dummies = pd.get_dummies(df,columns=['Profession',"Bank_dep","Risk","Number of credits"])
encode_dummies
```

```python
#One-hot encoding method
#we will not use on risk and number of credits because they are ordinal values
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
encoded = encoder.fit_transform(df.loc[:,["Profession","Bank_dep"]])
```

```python
df_encoded = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names(["Profession","Bank_dep"]))
df_encoded
```

```python
#ordinal
scale_mapper = {"High":3,"high":3, "medium":2, "low":1}
df["Scale_Risk"] = df["Risk"].replace(scale_mapper)
df
```

```python
df.drop(['Scale'], axis=1)
```
