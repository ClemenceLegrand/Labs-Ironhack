#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.arange(0,100)
y=x**2
z=np.exp(x)
fig, axes= plt.subplots(1,2)
axes[0].plot(x,y, color="#32a852",label="y function")
axes[0].plot(x,z, color="red",label="z function")
axes[0].set_xscale("linear")
axes[0].set_yscale("linear")
axes[1].plot(x,y, color="#32a852",label="y function")
axes[1].plot(x,z, color="red",label="z function")
axes[1].set_xscale("log")
axes[1].set_yscale("log")
```

#### As a bonus challenge, try to add a legend to the plot.

```python
x = np.arange(0,100)
y=x**2
z=np.exp(x)
fig, axes= plt.subplots(1,2)
axes[0].plot(x,y, color="#32a852",label="y function")
axes[0].plot(x,z, color="red",label="z function")
axes[0].set_xscale("linear")
axes[0].set_yscale("linear")
axes[1].plot(x,y, color="#32a852",label="y function")
axes[1].plot(x,z, color="red",label="z function")
axes[1].set_xscale("log")
axes[1].set_yscale("log")
axes[0].legend()
axes[1].legend()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_steps_month = fitbit.groupby(['Months_encoded'])['Steps'].median()
print(median_steps_month)
# MEDIAN STEPS BY WORK_OR_WEEKEND
median_work_weekend = fitbit.groupby(['Work_or_Weekend'])['Steps'].median()
median_work_weekend #work=1 weekend=0
```

```python
fig, axes = plt.subplots(1,2,figsize = (12,4))
axes[0].bar(median_steps_month.index, median_steps_month)
axes[0].set_title('Median Steps per Month', fontweight='bold')
axes[0].set_xticks(ticks=[1,3,5,7,9,11],labels=["january","march","may","july","sept","nov"]) 
axes[1].bar(median_work_weekend.index, median_work_weekend)
axes[1].set_title('Median Steps on Work days and Week end', fontweight='bold')
axes[1].set_xticks(ticks=[1,0],labels=["work","week-end"]) 
fig.savefig("figure1")
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
sns.histplot(x="Age",data=titanic, kde=True)
mean = np.mean(titanic["Age"])
plt.axvline(mean, color='red', label='mean')
median=np.median(titanic["Age"])
plt.axvline(median, color='green', label='median')
#var=np.var(titanic["Age"])
#plt.axvline(var, color='black', label='var')
q1 = np.percentile(titanic["Age"], 25)
q3 = np.percentile(titanic["Age"], 75)
plt.axvline(q1, color='black', label='Q1')
plt.axvline(q3, color='blue', label='Q3')
std=np.std(titanic["Age"])
plt.axvline(std, color='grey', label='std')
plt.legend()
```

```python
sns.set(style="darkgrid")
# creating a figure composed of two matplotlib.Axes objects (ax_box and ax_hist)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
# assigning a graph to each ax
sns.boxplot(data=titanic,x="Age", ax=ax_box)
sns.histplot(titanic, x="Age", ax=ax_hist)
f.tight_layout()
# Remove x axis name for the boxplot
ax_box.set(xlabel="")
```
