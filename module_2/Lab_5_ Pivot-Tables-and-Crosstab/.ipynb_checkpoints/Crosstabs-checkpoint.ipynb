### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(moovies.country,np.count_nonzero)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([moovies.country,moovies.genres,moovies.director_name],np.count_nonzero)
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(moovies.color,columns="Mean", values = moovies.duration,aggfunc="mean")
```
